**Task: Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.**

```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the dataset
file_path = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv'  # update if different
df = pd.read_csv(file_path, skiprows=4)

# Step 3: Select top 10 countries by population in 2023
df_2023 = df[['Country Name', '2023']].dropna()
top_countries = df_2023.sort_values(by='2023', ascending=False).head(10)

# Step 4: Plot the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_countries['Country Name'], top_countries['2023'], color='Purple', edgecolor='black')
plt.title('Top 10 Most Populous Countries in 2023')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Create a sample dataset for age distribution
data = {
    'Age Group': ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70+'],
    'Population': [15000, 18000, 22000, 20000, 17000, 14000, 10000, 7000]
}

df = pd.DataFrame(data)

# Step 3: Display the dataset
print("Sample Age Group Population Data:")
print(df)

# Step 4: Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['Age Group'], df['Population'], color='blue', edgecolor='black')
plt.title('Population Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Population')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Step 5: Show the plot
plt.show()
```
